```python
import torch
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# bhtrace components
from bhtrace.geometry import MinkowskiCart, Photon
from bhtrace.tracing import PTracer
```

## 1. Setup Spacetime and Particle

First, we define the spacetime geometry and the type of particle we want to trace. For this example, we'll use a simple `MinkowskiCart` (flat) spacetime and a `Photon`.

```python
# Define the spacetime
spacetime = MinkowskiCart()

# Define the particle
particle = Photon(spacetime)
```

## 2. Setup Tracer and Initial Conditions

Next, we instantiate a tracer. `PTracer` is a Hamiltonian ray-tracer. We also need to define the initial conditions for our particle(s): initial position `X0` and initial covariant momentum `P0`.

We will trace a single photon.
- **Initial Position (X0):** `[t=0, x=0, y=1, z=0]`
- **Initial Momentum (P0):** `[p_t=1, p_x=-1, p_y=0, p_z=0]` (moving towards negative x)

```python
# Instantiate the tracer. We'll use the RK4 integration method.
tracer = PTracer(ode_method='Euler')

# Initial position: (t, x, y, z)
X0 = torch.tensor([[0.0, 0.0, 1.0, 0.0]])

# Initial covariant momentum: (p_t, p_x, p_y, p_z)
P0 = torch.tensor([[1.0, -1.0, 0.0, 0.0]])

# Integration parameters
T = 1.0      # Total integration time
nsteps = 100 # Number of steps
```

## 3. Run the Tracer

Now we can call the `forward` method of the tracer to compute the trajectory.

```python
# Run the tracer
X, P = tracer.forward(particle, X0, P0, T, nsteps)

# The tracer returns the positions (X) and momenta (P) at each step.
print("Shape of X:", X.shape)
print("Shape of P:", P.shape)
```
